### Write a list comprehension such that even number should be replaced by 0 keep the odd number as it is.
- input[1,2,3,4]
- output[1,0,3,0]
   #given list
   l = range(1,12)

```python
l = range(1, 12)
print('original:', list(l))
output = [0 if x % 2 == 0 else x for x in l]
print('result:',output)
```

### write a program to reserve the string elements in the list
- input:['abc','def','ghi]
- output:['cba','fed','igh']

```python
def reverse_string(l):
    print(
        [ele[::-1] for ele in l]
    )

reverse_string(['abc','def','ghi'])
```
